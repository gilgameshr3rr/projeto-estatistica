Este notebook re칰ne etapas de explora칞칚o, organiza칞칚o e an치lise estat칤stica do conjunto de dados sobre vendas globais de jogos de videogame at칠 2024.

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/gilgameshr3rr/projeto-estatistica/project-edition-larissa/dataset/vgchartz-2024.csv'
df = pd.read_csv(url) 
df.head()
```

```python
import pandas as pd

# N칰mero total de registros (linhas) e vari치veis (colunas)
num_linhas, num_colunas = df.shape

resumo_tamanho = pd.DataFrame({
    'Descri칞칚o': ['Total de registros (linhas)', 'Total de vari치veis (colunas)'],
    'Quantidade': [num_linhas, num_colunas]
})

resumo_tamanho
```

```python
from IPython.display import Markdown, display

# Classifica칞칚o de vari치veis
quantitativas = []
qualitativas = []

for coluna in df.columns:
    if pd.api.types.is_numeric_dtype(df[coluna]):
        quantitativas.append(coluna)
    else:
        qualitativas.append(coluna)

display(Markdown(f"**游늵 Vari치veis Quantitativas:** {', '.join(quantitativas)}"))
display(Markdown(f"**游댟 Vari치veis Qualitativas:** {', '.join(qualitativas)}"))
```

```python
# Tabela resumo das vari치veis
resumo = pd.DataFrame({
    'Vari치vel': quantitativas + qualitativas,
    'Tipo': ['Quantitativa'] * len(quantitativas) + ['Qualitativa'] * len(qualitativas)
})

resumo
```
